```python
def is_palindrome(s):
    s = ''.join(char.lower() for char in s if char.isalnum())
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

test_string = "A man, a plan, a canal: Panama"
print("Is '{}' a palindrome?".format(test_string), is_palindrome(test_string))
```

```python
def find_duplicate_number(nums):
  seen=set()
  dupli=set()
  for num in nums:
    if num in seen:
      dupli.add(num)

    seen.add(num)
  return dupli

test_nums=[1,3,4,2,2,4]
print(find_duplicate_number(test_nums))
```

```python
def two_sum(nums,target):
  seen={}
  for i,num in enumerate(nums):
    complement=target - num
    if complement in seen:
      return[seen[complement],i]
    seen[num]=i

test_nums=[2, 7, 11, 15]
test_target = 6
print( two_sum(test_nums, test_target))
```

```python
def custom_sort(arr):
  for i in range(len(arr)):
    for j in range(i+1,len(arr)):
      if arr[i]>arr[j]:
        temp=arr[i]
        arr[i]=arr[j]
        arr[j]=temp
  return arr



arr = [5, 3, 9, 1, 7]
sorted_arr = custom_sort(arr)
print("Sorted array:", sorted_arr)
```

```python
def reverse_array(arr):
  start,end=0,len(arr)-1
  while start<end:
    temp=arr[start]
    arr[start]=arr[end]
    arr[end]=temp
    start+=1
    end-=1


my_array = [1, 2, 3, 4, 5]
reverse_array(my_array)
print("Reversed array:", my_array)
```

```python
def intersection(arr1,arr2,arr3):
  result=[]
  for num in arr1:
    if num in arr2 and num in arr3:
      result.append(num)
  return result


arr1 = [1, 2, 3, 4]
arr2 = [2, 3, 4, 5]
arr3 = [3, 4, 5, 6]
print("Intersection of three arrays:", intersection(arr1, arr2, arr3))
```

```python
def merge_sorted_arrays(arr1,arr2):
  result=[]
  i=j=0
  while i<len(arr1) and j< len(arr2):
    if arr1[i]< arr2[j]:
      result.append(arr1[i])
      i+=1
    else:
      result.append(arr2[j])
      j+=1
  result.extend(arr1[i:])
  result.extend(arr2[j:])
  return result



arr1 = [1, 3, 5,7,10,11]
arr2 = [2, 4, 6,9]
print("Merged sorted array:", merge_sorted_arrays(arr1, arr2))
```

```python
def max_sum(arr):
  max_sum=float('-inf')
  for i in range(len(arr)):
    for j in range(i,len(arr)):
      max_sum=max(max_sum,sum(arr[i:j+1]))
  return max_sum
arr = [1, -3, 5, -7, 2, 8, -4]
print("Maximum subarray sum (Brute force):", max_sum(arr))
```

```python
def sort_string(string):
  n=len(string)
  char_list=list(string)
  for i in range(n-1):
    for j in range(0,n-i-1):
      if char_list[j]>char_list[j+1]:
        temp= char_list[j]
        char_list[j]=char_list[j+1]
        char_list[j+1]=temp
  return ''.join(char_list)


string = "cba"
print(sort_string(string))
```

```python
def stringcompress(string):
  compressed=""
  count=1
  for i in range(1,len(string)):
    if string[i]==string[i-1]:
      count+=1
    else:
      compressed+=string[-1]+str(count)
  compressed+=string[-1]+str(count)
  return compressed




string = "aabcccccaaa"
print("Compressed string:", stringcompress(string))

```

```python
def longesteven(string):
    max_length=0
    longest_word=""
    words=string.split()
    for word in words:
      if len(word)%2==0 and len(word)>max_length:
        max_length=len(word)
        longest_word= max_length
    return longest_word

string = "This is a sample string with several words of different lengths"
print ("Longest even word:", longesteven(string)  )
```

```python
def cameltosnake(camel_case):
  snake_case=""
  for char in camel_case:
    if char[0]==char.isupper:
      snake_case+= char
    elif char.isupper():
      snake_case+="_"+char.lower()
    else:
      snake_case+= char
  return    snake_case


camel_case = "CamelCaseExample"
print("Snake case:", cameltosnake(camel_case))


```
